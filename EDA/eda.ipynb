## Килерфича (Сезонность)

```python
# Преобразуйте столбец с датой и временем в тип datetime
df_two_data['data_and_time'] = pd.to_datetime(df_two_data['data_and_time'], errors='coerce')

# Определите функцию для определения сезона
def sezon(data):
    if pd.notna(data):
        month = data.month
        if 3 <= month <= 5:
            return 'Весна'
        elif 6 <= month <= 8:
            return 'Лето'
        elif 9 <= month <= 11:
            return 'Осень'
        else:
            return 'Зима'
    else:
        return 0

# Примените функцию для создания столбца "Сезон"
df_two_data['seasonality'] = df_two_data['data_and_time'].apply(sezon)
```

## Килерфича (Время суток)

```python
# Преобразуйте столбец с датой и временем в тип datetime
df_two_data['data_and_time'] = pd.to_datetime(df_two_data['data_and_time'], errors='coerce')

# Определите функцию для определения времени суток
def time_of_day(data):
    hour = data.hour
    if 5 <= hour < 11:
        return 'Утро'
    elif 11 <= hour < 17:
        return 'День'
    elif 17 <= hour < 23:
        return 'Вечер'
    else:
        return 'Ночь'

# Примените функцию для создания столбца "Время суток"
df_two_data['time_of_day'] = df_two_data['data_and_time'].apply(time_of_day)

df_two_data
```
