```python
# Преобразование столбца 'data_and_time' в формат даты
df['data_and_time'] = pd.to_datetime(df['data_and_time'])

# Создание нового столбца 'date' с датой без времени
df['date'] = df['data_and_time'].dt.date

# Группировка данных по 'date', 'district', 'type' и подсчет количества происшествий
result = df.groupby(['date', 'district', 'type']).size().reset_index(name='count')

# Используем pivot_table для создания столбцов для каждого типа ЧС
pivot_result = result.pivot_table(index=['date', 'district'], columns='type', values='count', fill_value=0).reset_index()

# Объединяем с исходным столбцом 'seasonality'
final_result = pd.merge(pivot_result, df[['date', 'seasonality', 'district']], on=['date', 'district'])

final_result = final_result[['date', 'district', 'seasonality'] + final_result.columns.difference(['date', 'district', 'seasonality']).tolist()]

# Вывод результата
final_result
```

```python
import pandas as pd

# Группировка данных по 'date', 'district' и 'type' для подсчета, в какие дни произошли события
event_days = df.groupby(['date', 'district', 'type']).size().reset_index(name='count')

# Создание DataFrame, в котором каждая строка представляет день и вероятности для каждого типа события
result = pd.DataFrame()

# Подсчет вероятности для каждого типа события для каждого дня
for date in event_days['date'].unique():
    day_events = event_days[event_days['date'] == date]
    day_probability = day_events.groupby('type')['count'].count() / len(day_events)
    day_probability['date'] = date
    result = pd.concat([result, day_probability.to_frame().T])

# Вывод результатов
result
```

```python
# Загрузка датасет
df = pd.read_csv("Intermediate_datasets/merged_df_2.csv")

# Создание словарь с соответствиями старых и новых регионов
replace_dict = {
    "Снежный покров не постоянный.": "0",
    "Менее 0.5": "0.5",
    "Измерение невозможно или неточно.": "0",
}

# Замена старых регионов на новые
df["sss"] = df["sss"].replace(replace_dict)

df["sss"] = pd.to_numeric(df["sss"], errors="coerce")

# Сохранение измененного DataFrame в новый файл
df.to_csv("C:/Users/huawei/Desktop/Данные/merged_df_2.csv", index=False)
```
