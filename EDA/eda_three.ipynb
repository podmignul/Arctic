```python
df['RRR'] = df['RRR'].str.rstrip()

# Создайте словарь с соответствиями старых и новых регионов
replace_dict = {
    "Следы осадков": "0",
    "Осадков нет": "0",
}

# Замените старые регионы на новые
df["RRR"] = df["RRR"].replace(replace_dict)

df["RRR"] = pd.to_numeric(df["RRR"], errors="coerce")

df["RRR"]
```

```python
df['N'] = df['N'].str.rstrip()

# Создайте словарь с соответствиями старых и новых регионов
replace_dict = {
    "Облаков нет.": "0",
    "90  или более, но не 100%": "95",
    "70 – 80%.": "75",
    "100%.": "100",
    "20–30%.": "25",
    "40%.": "40",
    "60%.": "60",
    "50%.": "50",
    "Небо не видно из-за тумана и/или других метеорологических явлений.": "0",
    "10%  или менее, но не 0": "5",

}

# Замените старые регионы на новые
df["N"] = df["N"].replace(replace_dict)

df["N"] = pd.to_numeric(df["N"], errors="coerce")

df["N"]
```

```python
# Загрузка данных
df = pd.read_csv("Intermediate_datasets/новый_Данныепометеостанциям.csv")

# Преобразуйте столбец "Местное время" в формат даты и времени
df['Местное время'] = pd.to_datetime(df['Местное время'], format='%d.%m.%Y %H:%M')

# Определите день и дату для каждой записи
df['Дата'] = df['Местное время'].dt.date

# Разделите датасет по уникальным значениям "meteostation"
unique_meteostations = df['meteostation'].unique()

# Создайте пустой DataFrame для хранения итоговых данных
final_data = pd.DataFrame()

# Обработайте данные для каждой метеостанции и объедините результаты
for meteostation in unique_meteostations:
    meteostation_data = df[df['meteostation'] == meteostation]

    # Группировка данных по дням для текущей метеостанции
    daily_data = meteostation_data.groupby('Дата').agg({
        'T': 'mean',
        'Po': 'mean',
        'P': 'mean',
        'Pa': 'mean',
        'U': 'mean',
        'DD': 'first',
        'Ff': 'mean',
        'ff10': 'mean',
        'ff3': 'mean',
        'N': 'mean',
        'WW': 'first',
        'W1': 'first',
        'W2': 'first',
        'Tn': 'min',
        'Tx': 'max',
        'Cl': 'first',
        'Nh': 'mean',
        'H': 'first',
        'Cm': 'first',
        'Ch': 'first',
        'VV': 'first',
        'Td': 'mean',
        'RRR': 'mean',
        'tR': 'sum',
        'E': 'first',
        'Tg': 'mean',
        "E'": 'first',
        'sss': 'first',
        'meteostation': 'first'
    }).reset_index()

    # Добавьте данные для текущей метеостанции в итоговый DataFrame
    final_data = pd.concat([final_data, daily_data])

# Вывод нового DataFrame
final_data
```

```python
final_data = final_data[(final_data['meteostation'] != 'Кын') & (final_data['meteostation'] != 'Усть-Черная')]
final_data
```
